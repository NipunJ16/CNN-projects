```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv("../input/age-gender-and-ethnicity-face-data-csv/age_gender.csv")
df.head()
```

```python
df['pixels'] = df['pixels'].apply(lambda x:  np.array(x.split(), dtype="float32"))
```

```python
df['pixels'] = df['pixels'].apply(lambda x : x / 255)
```

```python
x = np.array(df['pixels'].tolist())
```

```python
x = x.reshape(x.shape[0], 48, 48, 1)
```

```python
y = df['gender']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization, InputLayer
```

```python
model = keras.Sequential([
    
    InputLayer(input_shape = (48, 48, 1)),
    
    Conv2D(32, (3, 3), padding = "same", input_shape = (32, 32, 3), activation = "relu"),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.2),
    
    Conv2D(64, (3, 3), padding = "same", activation = "relu"),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.2),
    
#     Conv2D(64, (3, 3), padding = "same", activation = "relu"),
#     BatchNormalization(),
#     MaxPooling2D((2, 2)),
#     Dropout(0.2),
    
    Flatten(),
    
    Dense(64, activation = "relu"),
    Dropout(0.4),
    
    Dense(1, activation = 'sigmoid')

])

model.summary()
```

```python
model.compile(
    optimizer = "adam",
    loss = "binary_crossentropy",
    metrics = ['accuracy']
)
```

```python
model_info = model.fit(x_train, y_train, epochs = 15, batch_size = 64, validation_split = 0.1)
```

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose = 0)
loss, accuracy
```

```python
model.save("model_gender.h5")
```
